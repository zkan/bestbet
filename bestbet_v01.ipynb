```python
import numpy as np
import pandas as pd
from scipy.stats import skewnorm  
%matplotlib inline
```

```python
starting_stack = 10000
current_stack = starting_stack
```

```python
current_stack = starting_stack
i = 1
while i < 51:
    #setting the base values
    
    #multiplier
    #beginner mode (10, 0.4, 2)
    #advanced mode
    multiplier = skewnorm.rvs (10, 0.2, 1.3)
    multiplier = round(multiplier,1)
    if multiplier < 1:
        continue

    #chance_of_winning
    mu = 0.47 #mean
    sigma = 0.16#stddev
    chance_of_winning = np.random.normal(mu, sigma)
    if chance_of_winning < 0 or chance_of_winning > 1:
        continue
        
    #expected_value
    expected_value = chance_of_winning * multiplier

    #sentence
    print('ROUND #',i)

    cow = "{:.0f}".format(chance_of_winning * 100)+'%'
    print('You have', current_stack, 'money!')
    print('You have', cow, 'chance to', str(multiplier) + '* your money!')
#    print('    (expected_value:', expected_value, ')')
    answer = input('How much of your money are you willing to risk?')
    answer = int(answer)

    if answer == 0:
        print('----------------')
        i = i + 1
        continue
    elif answer > current_stack:
        print('')
        print('You do not have this much money! LOL')
        print('----------------')
    else:
        #draw
        risked_money = int(answer)
        current_stack = current_stack - risked_money
        chance_of_losing = 1 - chance_of_winning
        outcome = np.random.choice([True, False], p=[chance_of_winning, chance_of_losing])
        result = risked_money * multiplier * outcome
        result = int(result)
        current_stack = current_stack + result
        if result == 0:
            print('')
            print('woops, you lost, you have this much money left: ', current_stack)
            print('----------------')
        else:
            print('')
            print('yay, you won, now you have this much money: ', current_stack)
            print('----------------')
        i = i + 1
        
    if current_stack < 0:
        print('')
        print('loser')
        break
```
